```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
from matplotlib.patches import ConnectionPatch
```

```python
df=pd.read_csv("global-energy-substitution.csv")
df
```

```python
df.drop(["Entity","Code"],axis=1,inplace=True)
df
```

```python
df.set_index("Year",inplace=True)
df
```

```python
df.columns
```

```python
df.columns[0].split(" ")[0]
```

```python
new_columns=[]
for column in df.columns.tolist():
    new_columns.append(column.split(" ")[0])
new_columns
```

```python
new_columns[4] = "Traditional biomass"
new_columns
```

```python
df.columns=new_columns
df
```

```python
#Compare energy consumption between 1900 and 1800
df.loc[1900].sum()/df.loc[1800].sum()
```

```python
#Compare energy consumption between 1900 and 2000
df.loc[2000].sum()/df.loc[1900].sum()
```

```python
#Compare energy consumption between 2000 and 2019
df.loc[2019].sum()/df.loc[2000].sum()
```

```python
#Share of aggregate coal, oil and gas consumption in 2019
df.loc[2019][["Coal","Oil","Gas"]].sum()/df.loc[2019].sum()
```

```python
colors=["limegreen","gray","red","blue","purple","orange","green","yellow","black","steelblue"]
plt.rcParams["font.size"] = 20

fig,ax=plt.subplots(figsize=(18,12))
ax=df.plot(kind="area",stacked=True,ax=ax,
       color=colors)
ax.set_ylabel("TWh")
ax.set_title("Global Energy Consumption from 1800 to 2019 \nSource: BP Statistical Review of World Energy 2021",
            fontdict={"weight":"bold", "size":25})
ax.legend(bbox_to_anchor=(1,1))

#plt.tight_layout()
plt.savefig("Historical global energy consumption.jpeg",
            dpi=300,
           bbox_inches = "tight")
plt.show()
```

**Tensorflow**

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```python
SAMPLES = 250
Z_MIN=-1
Z_MAX=1
Z2_MIN=-10
Z2_MAX=10

data = np.asarray([x for x in np.linspace(Z_MIN,Z_MAX,SAMPLES)])
data2 = np.asarray([x for x in np.linspace(Z2_MIN,Z2_MAX,SAMPLES)])
```

**baseline**

```python
plt.figure(figsize=(5,4))
plt.plot(data)
plt.title(f'Sample Size {SAMPLES}')
plt.xlabel('X')
plt.ylabel('Y')
```

**TF Activations**

```python
# activations:  mish, relu, gelu, elu, selu, linear, softplus, swish, exponential, tanh
data_p = tf.keras.activations.gelu(data)
data2_p = tf.keras.activations.gelu(data2)

plt.figure(figsize=(5,4))
plt.plot(data, data_p)
plt.title(f'Sample Size {SAMPLES}')
plt.xlabel('X')
plt.ylabel('Y')

plt.figure(figsize=(5,4))
plt.plot(data2, data2_p)
plt.title(f'Sample Size {SAMPLES}')
plt.xlabel('X')
plt.ylabel('Y')
```
